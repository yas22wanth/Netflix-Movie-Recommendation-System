```python
def fetch_cast_name(data):
    
    L = []
    counter = 0

    for i in ast.literal_eval(data):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    
    return L
```

```python
def fetch_director(data):
    List = []
    
    for i in ast.literal_eval(data):
        if i['job'] == 'Director':
            List.append(i['name'])
            
    return List
```

# Using some NLP Tasks

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    lst = []
    
    for i in text.split():
        lst.append(ps.stem(i))
        
    return " ".join(lst)
```

```python
stem(new_df['tags'][0])
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# Now make a final function to print top 5 similar movies

```python
def recommend(movie):
    
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x: x[1])[1:6]
    
    for i in movie_list:
#         print(i[0])
        print(new_df.iloc[i[0]].title)        
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```

```python
recommend('Batman')
```
